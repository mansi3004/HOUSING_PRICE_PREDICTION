# Heatmap plot 

```python
from matplotlib.colors import ListedColormap
plt.figure(figsize=(20,18))
#corr=train_dataset.corr()
corr=train_dataset.astype(float).corr()
sns.set(style='ticks',color_codes=True)
sns.heatmap(corr,linewidths=0.1,square=True,linecolor='white',annot=True)
```

From the above Heatmap we can observe that blessings and capital distance are closely related with correlation on -1 and 
knights house distance and capital distance with correlation of 1.

Also garden and kings visited are closely related with correlation of 1.

And also sorcerer cursed and holy tree are related with correlation of 0.96.

```python
sns.pairplot(train_dataset, x_vars='king_visited',y_vars='garden',size=7,aspect=0.7)
```

The two points in this plot shows that the king visited the houses that had garden. 

```python
sns.pairplot(train_dataset, x_vars='knights_house_distance',y_vars='capital_distance',size=7,aspect=0.7)
```

This plot shows that the knights house distance and capital ditance have a linear relation between them.

```python
sns.pairplot(train_dataset, x_vars='blessings',y_vars='capital_distance',size=7,aspect=0.7)
```

This plot shows that the capital distance and blessings are inversely correlated and have a linear relation with slope -1.
