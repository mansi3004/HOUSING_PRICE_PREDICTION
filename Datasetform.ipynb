This function reads the filename and open it after that data is extracted from that file

```python
def filecall(filename):
    file=open(filename,mode='r')
    file
    text=file.read()
    file.close()
    return text
```

```python
def array_form(text):
    myarray=[]
    for line in text.split('\n\n\n'):
        myarray.append(line)
    return myarray
```

Count the total no. of houses(n) in a file and prepared a numpy array with n row containing 20 columns set each value to -1
which indicates data is not filled

```python
def housecount(myarray):
    n=0
    for line in myarray:
        if line.find("House ID : ")!=-1:
            n=n+1
    print(n)
    return n
```

```python
def new(n):
    new_set=np.zeros((n,20))
    for i in range(0,n):
        for j in range(0,20):
            new_set[i][j]=-1
    return new_set
```

Data is extracted from text by splitting lines and finding some specific words in text also our data is available in string form so it is converted into some numerical data explicitly

```python
def data_form(myarray,new_set):
    i=-1
    i=-1
    for l in myarray:
        i=i+1
        for line in l.split('\n'):
            if line.find("House ID : ")!=-1:
                p=line.split(" :  ")[1]
                new_set[i][0]=int(p,16)
            if line.find("Date Built : ")!=-1:
                l=line.split(" and ")
                k=l[0]
                p=k.split(" : ")[1]
                s=p.split(" ")[0]
                a=s.split("/")
                new_set[i][1]=int(a[2])*365+int(a[0])*31+int(a[1])
                k=l[1]
                print(k)
                p=k.split(" :  ")[1]
                print(p)
                s=p.split(" ")[0]
                print(s)
                a=s.split("/")
                print(a)
                new_set[i][2]=int(a[2])*365+int(a[0])*31+int(a[1])
            if line.find("garden")!=-1:
                if line.find("no")!=-1:
                    new_set[i][3]=0
                else:
                    new_set[i][3]=1
            if line.find("Dock")!=-1:
                p=re.findall('[0-9.]+',line)
                new_set[i][4]=float(p[0])
            if line.find("Capital")!=-1:
                p=re.findall('[0-9.]+',line)
                new_set[i][5]=float(p[0])
            if line.find("Royal Market ")!=-1:
                p=re.findall('[0-9.]+',line)
                new_set[i][6]=float(p[0])
            if line.find("Guarding Tower ")!=-1:
                p=re.findall('[0-9.]+',line)
                new_set[i][7]=float(p[0])
            if line.find("River")!=-1:
                p=re.findall('[0-9.]+',line)
                new_set[i][8]=float(p[0])
            if line.find("renovation")!=-1:
                if line.find("underwent")!=-1:
                    new_set[i][9]=1
                else:
                    new_set[i][9]=0
            if line.find("dining rooms")!=-1:
                p=re.findall('[0-9]+',line)
                new_set[i][10]=int(p[0])
            if line.find("bedrooms")!=-1:
                p=re.findall('[0-9]+',line)
                new_set[i][11]=int(p[0])
            if line.find("bathrooms")!=-1:
                p=re.findall('[0-9]+',line)
                new_set[i][12]=int(p[0])
            if line.find("King Visited")!=-1: 
                new_set[i][13]=1
            if line.find("couldn't pay his visit ")!=-1: 
                new_set[i][13]=0
            if line.find("blessed")!=-1:
                p=re.findall('[0-9]+',line)
                new_set[i][14]=int(p[0])
            if line.find("small land ")!=-1: 
                new_set[i][15]=0.5
            if line.find("huge land ")!=-1: 
                new_set[i][15]=1    
            if line.find("no land ")!=-1: 
                new_set[i][15]=0
            if line.find("Location of the house ")!=-1:
                if line.find("Mountains")!=-1:
                    new_set[i][16]=0
                if line.find("Cursed")!=-1:
                    new_set[i][16]=1
                if line.find("Servant")!=-1:
                    new_set[i][16]=2
                if line.find("King")!=-1:
                    new_set[i][16]=3
            if line.find("Knight's house ")!=-1:
                p=re.findall('[0-9.]+',line)
                new_set[i][17]=float(p[0])
            if line.find("orcerer ")!=-1:
                if line.find("could")!=-1:
                    new_set[i][18]=0
                if line.find("cursed")!=-1:
                    new_set[i][18]=1
            if line.find("Holy tree")!=-1:
                if line.find("death")!=-1:
                    new_set[i][19]=0
                if line.find("tall")!=-1:
                    new_set[i][19]=1
    return new_set
            
```

This function converts numpy array formed from the file into a new dataframe and append it to total existing dataframe

```python
def datasetform(dataframe,new_set,columns,k):
    dfnew=pd.DataFrame(new_set,columns=columns)
    dfnew['builder']=k
    dataframe=dataframe.append(dfnew)
    return dataframe
```

To extract data from 11 files it became convenient to make a function tht takes filename and existing datafame so as to add data from file into existing dataframe

```python
def total_func(dataframe,filename,columns,k):
    text=filecall(filename)
    myarray=array_form(text)
    n=housecount(myarray)
    new_set=new(n)
    new_set=data_form(myarray,new_set)
    dataframe=datasetform(dataframe,new_set,columns,k)
    return dataframe
```
