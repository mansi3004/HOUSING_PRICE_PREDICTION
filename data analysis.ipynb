```python
train['bedrooms'].unique()
train.groupby('bedrooms').count()
```

```python
train[['builder', 'Golden Grains']].groupby(['builder'], as_index=False).mean().sort_values(by='Golden Grains', ascending=False)
```

```python
train[['location', 'Golden Grains']].groupby(['location'], as_index=False).nunique().sort_values(by='Golden Grains', ascending=False)
```

#this analysis provides the information that in most of the cases houses were near mountains

```python
train[['king_visited', 'Golden Grains']].groupby(['king_visited'], as_index=False).nunique().sort_values(by='Golden Grains', ascending=False)
```
